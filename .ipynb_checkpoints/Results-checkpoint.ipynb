```python
def evaluate(val_list, num, name):
    avg = np.mean(val_list)
    std = np.std(val_list)
    med = np.median(val_list)
    
    print("for number of vehicles = " + str(num) + '  '+ name)
    print("mean: " + str(avg))
    print("median: " + str(med))
    print("std: " + str(std))
    print()
    
    return avg,med,std

num_vehicles = [20,30,40,50]
stats = np.zeros([5,4,3])

for ind,num in enumerate(num_vehicles):
    bsl_f = glob.glob('base*%d.json'%num)[0]
    tst_f = glob.glob('with*%d.json'%num)[0]
    stg_f = glob.glob('strat*%d.json'%num)[0]
    nolc_f = glob.glob('no_lc*%d.json'%num)[0]
    dist_2_f = glob.glob('./inverse_distance_square/distance_square%d_range300.json'%num)[0]

    print(bsl_f,tst_f,stg_f,nolc_f)
    
    with open(bsl_f) as f:
        bsl = json.load(f)['episode_reward']
    with open(tst_f) as f:
        tst =  json.load(f)['episode_reward']
    with open(stg_f) as f:
        stg = json.load(f)['returns']
    with open(nolc_f) as f:
        no_lc = json.load(f)['returns']
    with open(dist_2_f) as f:
        dist2 = json.load(f)['episode_reward']

    avg,med,std = evaluate(bsl,num,"baseline")
    stats[0,ind,:] = np.array([avg,std,med])

    avg,med,std = evaluate(tst,num,"test")
    stats[1,ind,:] = np.array([avg,std,med])
    
    avg,med,std = evaluate(stg,num,"strategic")
    stats[2,ind,:] = np.array([avg,std,med])
    
    avg,med,std = evaluate(no_lc, num, "no lane changing")
    stats[3,ind,:] = np.array([avg,std,med])
    
    avg,med,std = evaluate(dist2, num, "inverse distance square")
    stats[4,ind,:] = np.array([avg,std,med])
```

```python
model_ind_list = [1,4,0,2,3]
model_names = ['Linear Weighted DSQ','Quadratic Weighted DSQ','Unweighted DSQ','Rule Based (LC2013)','No lane Changing']
color_list = ['tomato','dodgerblue','limegreen','peru','grey']
num_vehicles = [20,30,40,50]

def plot_mean_with_std(ax,stats,model_ind,cl,lb):
    ax.plot(np.array(num_vehicles),stats[model_ind,:,0],label=lb,color=cl)
    
    CI = 1.96 * stats[model_ind,:,1]/np.sqrt(10)
    
    bound_upper = stats[model_ind,:,0]+CI
    bound_lower = stats[model_ind,:,0]-CI
    
    print(bound_lower, bound_upper)
    ax.fill_between(np.array(num_vehicles),bound_upper,bound_lower,color=cl,alpha=0.2)
    ax.set_ylim(750,1700)
    ax.set_xlabel("Number of vehicles in scenario",fontsize=20)
    ax.set_ylabel("Reward",fontsize=20)
    
    
def plot_median(ax,stats,model_ind,cl,lb):
    ax.plot(np.array(num_vehicles),stats[model_ind,:,2],label=lb,color=cl)
    ax.set_xlabel("Number of vehicles in scenario",fontsize=20)
    ax.set_ylabel("Reward",fontsize=20)
    
fig, ax = plt.subplots(figsize=(10,8),dpi=100)
for i in range(len(model_names)):
    plot_mean_with_std(ax,stats,model_ind_list[i],color_list[i],model_names[i])
ax.tick_params(labelsize=20)
ax.set_title('mean',fontsize=20)
plt.legend(prop={'size': 15})
plt.savefig('/Users/jiqiandong/Desktop/OneDrive - purdue.edu/Jiqian\'s work/draft/figures/mean2.png',dpi=100)
plt.show()

```

```python
from scipy.optimize import curve_fit
import matplotlib.pylab as plt
import numpy as np


def generate_figure(x_data,y_data,t=False):
    plt.figure(figsize=(10,6),dpi=100)
    func = lambda x, a, b: a * (1 / (x)**2) + b 
#     func = lambda x,a,b,c,d: a * x**3 + b*x**2 + c*x+ d
    
    # averages and dispersion of data
    y_means = np.array(y_data).mean(axis = 1)
    y_min = np.array(y_data).min()
    y_max = np.array(y_data).max()
    y_spread = np.array(y_data).std(axis = 1)

    best_fit_ab, covar = curve_fit(func, x_data, y_means,
                                   sigma = y_spread,
                                   absolute_sigma = True)
    sigma_ab = np.sqrt(np.diagonal(covar))

    from uncertainties import ufloat
    a = ufloat(best_fit_ab[0], sigma_ab[0])
    b = ufloat(best_fit_ab[1], sigma_ab[1])
    text_res = "Best fit parameters:\na = {}\nb = {}".format(a, b)
    print(text_res)

    # plotting the unaveraged data
    
    
#     flier_kwargs = dict(marker = 'o', markerfacecolor = 'silver',
#                         markersize = 3, alpha=0.7)
#     line_kwargs = dict(color = 'k', linewidth = 1)
    bp = plt.boxplot(y_data, positions = x_data,
#                      patch_artist=True,
#                      capprops = line_kwargs,
#                      boxprops = line_kwargs,
#                      whiskerprops = line_kwargs,
#                      medianprops = line_kwargs,
#                      flierprops = flier_kwargs,
                     widths = 20,
                     meanline=True,
                     manage_xticks = False)
    
    
#     plotting the averaged data with calculated dispersion
#     plt.scatter(x_data, y_means, facecolor = 'silver', alpha = 1)
#     plt.errorbar(x_data, y_means, y_spread, fmt = 'none', ecolor = 'black')

    # plotting the model
    hires_x = np.linspace(100, 500, 50)
    plt.plot(hires_x, func(hires_x, *best_fit_ab))
    bound_upper = func(hires_x, *(best_fit_ab + sigma_ab))
    bound_lower = func(hires_x, *(best_fit_ab - sigma_ab))
    # plotting the confidence intervals
    plt.fill_between(hires_x, bound_lower, bound_upper,
                      alpha = 0.15)
    plt.xlim(50, 550)
    if t:
        plt.ylim(y_min*0.95, 1900)
    else:
        plt.ylim(y_min*0.95, y_max)
        
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.xlabel("connectivity range",fontsize = 20)
    plt.ylabel("reward",fontsize = 20)
    
    
    
    
num_vehicles = [20,30,40,50]
connectivity_ranges = [100,200,300,400,500]

x_data = np.array(connectivity_ranges)
for t,num in enumerate(num_vehicles):
    y_data = []
    for conn in connectivity_ranges:
        tst_f = glob.glob('./connectivity_range_results/*%d_range%d.json'%(num,conn))[0]
        with open(tst_f) as f:
            tst = json.load(f)['episode_reward']
        y_data.append(np.array(tst))
    if t == 0:
        generate_figure(x_data,y_data,1)
    else:
        generate_figure(x_data,y_data)
    plt.title("number of vehicles: "+str(num),fontsize = 20)
    plt.savefig("./figures/connect_range_with_"+str(num)+'.png',dpi=300)
    plt.show()
    
```

```python
from scipy.optimize import curve_fit
import matplotlib.pylab as plt
import numpy as np


def generate_figure(x_data,y_data,t=False):
    plt.figure(figsize=(10,6),dpi=100)
    func = lambda x, a, b: a * (1 / (x)**2) + b 
#     func = lambda x,a,b,c,d: a * x**3 + b*x**2 + c*x+ d
    
    # averages and dispersion of data
    y_means = np.array(y_data).mean(axis = 1)
    y_min = np.array(y_data).min()
    y_max = np.array(y_data).max()
    y_spread = np.array(y_data).std(axis = 1)

    best_fit_ab, covar = curve_fit(func, x_data, y_means,
                                   sigma = y_spread,
                                   absolute_sigma = True)
    sigma_ab = np.sqrt(np.diagonal(covar))

    from uncertainties import ufloat
    a = ufloat(best_fit_ab[0], sigma_ab[0])
    b = ufloat(best_fit_ab[1], sigma_ab[1])
    text_res = "Best fit parameters:\na = {}\nb = {}".format(a, b)
    print(text_res)

    # plotting the unaveraged data
    
    
#     flier_kwargs = dict(marker = 'o', markerfacecolor = 'silver',
#                         markersize = 3, alpha=0.7)
#     line_kwargs = dict(color = 'k', linewidth = 1)
    bp = plt.boxplot(y_data, positions = x_data,
#                      patch_artist=True,
#                      capprops = line_kwargs,
#                      boxprops = line_kwargs,
#                      whiskerprops = line_kwargs,
#                      medianprops = line_kwargs,
#                      flierprops = flier_kwargs,
                     widths = 20,
                     meanline=True,
                     manage_xticks = False)
    
    
#     plotting the averaged data with calculated dispersion
#     plt.scatter(x_data, y_means, facecolor = 'silver', alpha = 1)
#     plt.errorbar(x_data, y_means, y_spread, fmt = 'none', ecolor = 'black')

    # plotting the model
    hires_x = np.linspace(100, 500, 50)
    plt.plot(hires_x, func(hires_x, *best_fit_ab))
    bound_upper = func(hires_x, *(best_fit_ab + sigma_ab))
    bound_lower = func(hires_x, *(best_fit_ab - sigma_ab))
    # plotting the confidence intervals
    plt.fill_between(hires_x, bound_lower, bound_upper,
                      alpha = 0.15)
    plt.xlim(50, 550)
    if t:
        plt.ylim(y_min*0.95, 1900)
    else:
        plt.ylim(y_min*0.95, y_max)
        
    plt.xticks(fontsize=20)
    plt.yticks(fontsize=20)
    plt.xlabel("connectivity range",fontsize = 20)
    plt.ylabel("reward",fontsize = 20)
    
    
    
    
num_vehicles = [20,30,40,50]
connectivity_ranges = [100,200,300,400,500]

x_data = np.array(connectivity_ranges)
for t,num in enumerate(num_vehicles):
    y_data = []
    for conn in connectivity_ranges:
        tst_f = glob.glob('./connectivity_range_results2/base*%d_range%d.json'%(num,conn))[0]
        with open(tst_f) as f:
            tst = json.load(f)['episode_reward']
        y_data.append(np.array(tst))
    if t == 0:
        generate_figure(x_data,y_data,1)
    else:
        generate_figure(x_data,y_data)
    plt.title("number of vehicles: "+str(num),fontsize = 20)
    plt.savefig("./figures/baseline_connect_range_with_"+str(num)+'.png',dpi=300)
    plt.show()
    
```

```python
from scipy.optimize import curve_fit
import matplotlib.pylab as plt
import numpy as np

# model function
func = lambda x, a, b: a * (1 / (x**2)) + b 

# approximating OP points
n_ypoints = 7 
x_data = np.linspace(70, 190, n_ypoints)

# approximating the original scatter in Y-data
n_nested_points = 1000
point_errors = 50
y_data = [func(x, 4e6, -100) + np.random.normal(x, point_errors,
          n_nested_points) for x in x_data]

# averages and dispersion of data
y_means = np.array(y_data).mean(axis = 1)
y_spread = np.array(y_data).std(axis = 1)

best_fit_ab, covar = curve_fit(func, x_data, y_means,
                               sigma = y_spread,
                               absolute_sigma = True)
sigma_ab = np.sqrt(np.diagonal(covar))

from uncertainties import ufloat
a = ufloat(best_fit_ab[0], sigma_ab[0])
b = ufloat(best_fit_ab[1], sigma_ab[1])
text_res = "Best fit parameters:\na = {}\nb = {}".format(a, b)
print(text_res)

# plotting the unaveraged data
flier_kwargs = dict(marker = 'o', markerfacecolor = 'silver',
                    markersize = 3, alpha=0.7)
line_kwargs = dict(color = 'k', linewidth = 1)
bp = plt.boxplot(y_data, positions = x_data,
                 capprops = line_kwargs,
                 boxprops = line_kwargs,
                 whiskerprops = line_kwargs,
                 medianprops = line_kwargs,
                 flierprops = flier_kwargs,
                 widths = 5,
                 manage_xticks = False)
# plotting the averaged data with calculated dispersion
#plt.scatter(x_data, y_means, facecolor = 'silver', alpha = 1)
#plt.errorbar(x_data, y_means, y_spread, fmt = 'none', ecolor = 'black')

# plotting the model
hires_x = np.linspace(50, 190, 100)
plt.plot(hires_x, func(hires_x, *best_fit_ab), 'black')
bound_upper = func(hires_x, *(best_fit_ab + sigma_ab))
bound_lower = func(hires_x, *(best_fit_ab - sigma_ab))
# plotting the confidence intervals
plt.fill_between(hires_x, bound_lower, bound_upper,
                 color = 'black', alpha = 0.15)
plt.text(140, 800, text_res)
plt.xlim(40, 200)
plt.ylim(0, 1000)
plt.show()
```
